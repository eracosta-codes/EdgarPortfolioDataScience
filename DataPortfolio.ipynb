The First dataset being used the the Banking Collections' dataset.
The goal of this is to analyze the reasons why people have taken loans, how long is the credit for, how much credit people owe(total, and outstanding), the account types of the people taking loans, the customer's region, and risk levels based on a customer's score. This is important for banks as depending on collateral, a person's payment consistency, and the customer's credit score as these factors can determine a bank's confidence on their consumers.

The link to this dataset can be found here:

https://www.kaggle.com/datasets/kotich/banking-collections-dataset-synthetic-data 

Before each visual, there will be a brief description on what is being presented in the visual, otherwise, the visuals should be self-explanatory.

Countplot for users and their payment status


```python
fig, ax = plt.subplots()

skill = ['Missed', 'Partially Paid', 'Paid']
counts = [345, 330, 325]
bar_labels = ['red', 'blue' , 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:purple']

ax.bar(skill, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Users')
ax.set_title('Users and their Payment Status')
plt.show()
```

Countplot grouped by the Payment Status of users and their Risk Level with an order by Risk Levels

Categorical Bivariate data is being used as the payment level of different risk levels are being shown in order of risk levels

```python
risk_category = ("Low", "Medium", "High")
payment_status = {
    'Missed': (96, 130, 119),
    'Partially Paid': (91, 116, 118),
    'Paid': (22, 20, 22)
}

x = np.arange(len(risk_category))  # the label locations
width = 0.15  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in payment_status.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Users')
ax.set_title('User\'s Risk by Payment Status')
ax.set_xticks(x + width, risk_category)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 170)

plt.show()
```

```python
w_table = table1.loc[(table1["Region"] =="West")]

w_array = (np.array(w_table['Customer_Score'])).tolist()
```

```python
e_table = table1.loc[(table1["Region"] =="East")]

e_array = (np.array(e_table['Customer_Score'])).tolist()
```

```python
n_table = table1.loc[(table1["Region"] =="North")]

n_array = (np.array(n_table['Customer_Score'])).tolist()
```

```python
s_table = table1.loc[(table1["Region"] =="South")]

s_array = (np.array(s_table['Customer_Score'])).tolist()
```

By median, users who live in the southern region have more higher customer scores than every other geographic region

The data is split by region to explore their distribution of customer scores

Numerical/Categorical Bivariate data is being used as the numerical distributions of customer scores by region is being displayed

```python
geographic_regions = [
    w_array,
    e_array,
    n_array,
    s_array
]
labels = ['West', 'East', 'North', 'South']
colors = ['red', 'blue', 'purple', 'orange']

fig, ax = plt.subplots()
ax.set_ylabel('Customer Scores')
ax.set_title("Customer Scores by region")

bplot = ax.boxplot(geographic_regions,
                   patch_artist=True,  # fill with color
                   tick_labels=labels)  # will be used to label x-ticks

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
p_table = table1.loc[(table1["Payment_Status"] =="Paid")]

p_array = (np.array(p_table['Customer_Score'])).tolist()

pap_table = table1.loc[(table1["Payment_Status"] =="Partially Paid")]

pap_array = (np.array(pap_table['Customer_Score'])).tolist()

m_table = table1.loc[(table1["Payment_Status"] =="Missed")]

m_array = (np.array(m_table['Customer_Score'])).tolist()
```

By median, users who have fully paid their loans have more higher customer scores than every other payment status

The data is split by payment to explore their distribution of customer scores

Numerical/Categorical Bivariate data is being used as the numerical distributions of customer scores by payment status is being displayed

```python
payment_array = [
    p_array,
    pap_array,
    m_array
]
labels = ['Paid', 'Partially Paid', 'Missed']
colors = ['red', 'blue', 'purple']

fig, ax = plt.subplots()
ax.set_ylabel('Customer Scores')
ax.set_title("Customer Scores by payment status")

bplot = ax.boxplot(payment_array,
                   patch_artist=True,  # fill with color
                   tick_labels=labels)  # will be used to label x-ticks

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
score = (np.array(table1['Customer_Score'])).tolist()

emi = (np.array(table1['EMI_Amount'])).tolist()
```

There is no correlation between customer score and EMI Amount

A Scatterplot is being used to outline correlation between customer score and EMI Amount

Numerical/Numerical Bivariate data is being used as correlation is being used to outline correlation between an customer score and EMI amount

```python
sns.regplot(x=score, y=emi, ci=None, line_kws={"color": "red"})
plt.xlabel('Customer Score')
plt.ylabel('EMI Amount')
plt.title('Comparison of Customer Score versus EMI Amount')
plt.show()
```

There is no correlation between customer score and Outstanding Amount

A Scatterplot is being used to outline correlation between customer score and Outstanding Amount

Numerical/Numerical Bivariate data is being used as correlation is being used to outline correlation between an customer score and Outstanding amount

```python
outstanding_amt = (np.array(table1['Outstanding_Amount'])).tolist()
sns.regplot(x=score, y=outstanding_amt, ci=None, line_kws={"color": "red"})
plt.xlabel('Customer Score')
plt.ylabel('Outstanding Amount')
plt.title('Comparison of Customer Score versus Outstanding Amount')
plt.show()
```

There is a slight negative correlation between customer score and Loan Amount

Numerical/Numerical Bivariate data is being used as correlation is being used to outline correlation between an customer score and Loan amount

```python
loan_amt = (np.array(table1['Loan_Amount'])).tolist()
sns.regplot(x=score, y=loan_amt, ci=None, line_kws={"color": "red"})
plt.xlabel('Customer Score')
plt.ylabel('Loan Amount')
plt.title('Comparison of Customer Score versus Loan Amount')
plt.show()
```
